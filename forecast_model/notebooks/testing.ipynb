Predictors
- Event counts from the previous month in neighboring administrative divisions
- Interactions between different types of actors
- Fatality rates
- Strategic development events, such as whether a peace agreement was signed within the last six months
- Population estimates from WorldPop
- Infant mortality rates from the Center for International Earth Science Information Network

Data
- Data since 2018 are first split by country and a single country is selected
- ACLED CAST structures the data at the first administrative division (ADMIN1)-month level
- All subnational results per country are aggregated to produce the country-level prediction

Models
- Four types of models are fit to the data for each outcome type
  - (1) Battles, (2) explosions/remote violence, and (3) violence against civilians events
  - A random forest model and an xgboost model
  - Each using a “simple” and “complex” list of predictors

Cross Validation
-  The training data are further partitioned
into four time-series cross-validation splits, which sequentially hold out last six months of data
- Select the best-performing model
for each country based upon its predictive performance in the held-out data
- Fit a final model for each country using the full data since 2018, with no held-out data
